理論上、2になるはずだが、そうなっていない。   10進数でおおよそ17桁までの精度らしい。

```python
s = 0
while s != 1.:
    if s > 1.1:
        break
    print(s)
    s += 0.1
```

このように、意図しない挙動になることがあるので、注意が必要。  
回避策として、十分に小さな数epsを設定し、それを判定基準に採用する。

```python
eps = 1e-10
s = 0
# なぜこうしない？
#while s >= 1:
#    print(s)
#    s += 0.1
while abs(s-1) > eps:
    print(s)
    s += 0.1
```
